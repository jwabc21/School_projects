## Reading the file

Getting to know the file and what sort of observations you can take from the first couple entries can give a good idea of what to be looking for. For example, in this sample data, the ranking and index overlap and that is not needed, also in the first 30 entries there are various duplicates. Without this process, we wouldn't even know what about the data should be imputted or filtered.

```python
path = ''
df_sales = pd.read_csv(path + 'HW4-video-game-sales.csv')
df_sales.index = df_sales['Rank'] # set index
del df_sales['Rank'] # delete column once index has been set
df_sales.head(30)
```

```python
#First 100 entries
first = df_sales.loc[:100]
print(f"The % of entries that are intact is {100-first.isnull().sum().sum():.2f}%")
```

The justification for this approach is that we can see what values is null and take the sum of those values from the first 100 entries. The null/na rows is the dirtiest data, and it is difficult to individually see which data is fully intact if a lot of data is being processed. It is simpler to filter out the data that is completely blank and use the rest.

## Cleaning Data

Now that we know what data we might be looking at, it is time to check if there are human errors because this can influence the data analyzing process. Removing a row of all missing and dropping all NA's is what is needed to begin to understand how to analyze the data. 

```python
#filter where category might be missing
df_sales = df_sales[df_sales.Name != "Missing"]
#filter where there are at least 7 valid values
df_sales.dropna(thresh = 7, inplace = True)
df_sales.head(10)
```

## Creating New Data

To come up with meaningful insights, we need to create new columns that the original data might have missed. In this instance a total global sales column is added to see how well the game performs in all regions. 

```python
#converting data types to floats
df_sales['NA_Sales'] = pd.to_numeric(df_sales['NA_Sales'])
df_sales['EU_Sales'] = pd.to_numeric(df_sales['EU_Sales'])
df_sales['JP_Sales'] = pd.to_numeric(df_sales['JP_Sales'])
df_sales['Other_Sales'] = pd.to_numeric(df_sales['Other_Sales'])
df_sales['Year'] = pd.to_numeric(df_sales['Year'])
```

```python
df_sales["Total_Global_Sales"] = df_sales["NA_Sales"] + df_sales["EU_Sales"] + df_sales["JP_Sales"] + df_sales["Other_Sales"]
df_sales.head(10)
```

```python
df_sales.sort_values(by = "Total_Global_Sales", ascending = False).loc[:,['Name','Publisher:Platform','Total_Global_Sales']].head(15)
```

## User Friendly and Clean Data

We can see the human error of publisher and platform being put together. To create more variability for our data and allow for better analysis, we can seperate these two into two different categories.

```python
# Splitting 'Publisher:Platform' column into 'Publisher' and 'Platform'
df_sales[['Publisher', 'Platform']] = df_sales['Publisher:Platform'].str.split(':', n=-1, expand=True)
#drop the original column
df_sales.drop(columns = 'Publisher:Platform', inplace= True)
```

## Analysis of Games and what is Popular

Throughout this next section, we can see what attributes might make a game popular. We can see what companies are successful, which markets are attracting, and what names might be good for a game to be called by following popular sales. This allows for potential replication and a way to base a client recommendation on.

```python
# Number of nintendo games published
publisher = df_sales['Publisher']
print("The number of video games published by Nintendo is",publisher.isin(['Nintendo']).sum())
```

```python
# number of games that contain the name Super
print(df_sales['Name'].str.contains('Super', case=True).value_counts())
print("The number of games that contain Super in their name is 385.")
```

```python
#How many video games have reached sales of at least $20 million in at least one market?
# Filtering the DataFrame where any market sales are at least $20 million
success_sale = df_sales[(df_sales[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']] >= 20).any(axis=1)]
num_games = len(success_sale)
print(f"The number of video games with at least $20 million in sales in at least one market is {num_games}")
```

```python
#Games created past the year 2002
filtered = df_sales[df_sales['Year'] >= 2002]

#Check which of these games earned at least $2,000,000
successful_games = filtered[filtered['Total_Global_Sales'] >= 2]

#Calculate the percentage
percent = (len(successful_games) / len(filtered)) * 100

print(f"The percentage of video games created in 2002 and beyond that earned at least $2,000,000 in global sales is {percent:.2f}%")
```

## Correlations Between Markets and Total Global Sales

Observation 1
The correlation between NA_Sales and Global_Sales is very high, indicating that sales in North America are a strong predictor of global sales figures. This suggests that North America might be the largest market, or at least the most indicative of global success for these video games.

Observation 2
The correlation between JP_Sales and Global_Sales is lower compared to North America and Europe. This indicates that while Japanese sales are positively correlated with global sales, they are not as strong a predictor of global sales as North American or European sales.
