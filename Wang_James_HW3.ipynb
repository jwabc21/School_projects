## Displaying Data through a Function

This is a way to display the data when it has many different prompts. It can filter by column, value, and the number of rows you want to see in the data, sorting the data by profit highest to lowest. This can be changed to fit certain requirements, such as finding the lowest returns/negative profits. We explore several instances such as pulling a certain number of top profits based off a certain category. Business can utilize this to answer questions about their performance with certain orders and see if there is a relationship with a certain category and its success or failure.

```python
#function that returns sorted data
def data_table(df, col=None, val=None, n=15):
    #no value is given for column and value
    if col is None and val is None:
        #sorted data by profit and the number of desired rows
        return df.sort_values(by='Profit', ascending=False).head(n)
    #values given for column and value to search for
    else:
        #Search data for match by column and value
        filter_df = df[df[col] == val]
        #sorts the previous data
        sorted_data = filter_df.sort_values(by='Profit', ascending=False)
        return sorted_data.head(n)
```

```python
print("The top seven orders are displayed below.")
data_table(order_df,n=7)
```

```python
print("The top 10 orders in the state of Maryland are displayed below.")
data_table(order_df,col = "State",val = "Maryland",n=10)
```

```python
print("The top 20 orders by consumer segment are displayed below.")
data_table(order_df, col = "Segment", val = "Consumer",n=20)
```

```python
print("The top 5 orders with the zip code of 20852 are displayed below.")
data_table(order_df, col = "Postal Code", val = 20852,n=5)
```

## Determining Statistics

Statistics can easily be one of the easiest ways to find out answers about the orders. Knowing which are the best and worst performers and the profits associated with each is useful for maximizing profit and minimizing losses. However, the average and median of the data can also reveal where the business can work on to hopefully raise the numbers. If there are larger profits on each order on average, there could be potentially larger revenue gains.

```python
#Looking at Profit only
profit = order_df['Profit']
#minimum of profits
print(f'The biggest loss is ${profit.min():.2f}')
#maximum of profits
print(f'The biggest profit is ${profit.max():.2f}')
#average profit
print(f'The average profit is ${profit.mean():.2f}')
#median profit
print(f'The median profit is ${profit.median():.2f}')
```

## Capturing Market Share

Determining the segment of customer base the business make the most and least sales off of allows the business to figure out what they need to market to further strengthen or improve their customers. In this business, binders, paper, and furnishings are the most popular orders, which allows us to imply that offices or schools would be the major customer base. Their top 3 performers capture approximately 38.52% of their customers, which is a significant amount. If possible, there can be more of an even distribution for all their orders through certain business processes to keep a consistent revenue stream.

```python
#seperate subcategories from all orders
sub_categories_df = order_df['Sub-Category']
#display subcategory and count for interpretation
print(sub_categories_df.value_counts().head(3))
print("The three highest subcategories that the most orders originated in is Binders, Paper, and Furnishings.")
#make the counts into percent of orders
percent = sub_categories_df.value_counts()/len(order_df)*100
#display top 3 performers and their percentage of orders
print("Percentages",percent.head(3))
print("Binders make up 15%, paper makes up 13%, and furnishings make up 9% of orders.")

#Determine total percentage
top_three = percent['Binders':'Furnishings']
print(f'The total percentage of the top three sub-categories across all orders is {top_three.sum():.2f}%')
```

### Determining Outsourcing Location

```python
#Calculate # of orders per region
region = order_df['Region']
print(region.value_counts().head(10))
# Calculate average profit for each region
unique = order_df['Region'].unique()
avg_profit_per_region = {}
for region in unique:
    avg_profit_per_region[region] = order_df[order_df['Region'] == region]['Profit'].mean()
print(avg_profit_per_region)
order_df[order_df['Region'] == "West"]['State'].value_counts()
```

The best place to put the distibution center is in the west because it has both the highest average profits throughout the region and it generates the most sales out of all of the regions. This way it can bring in the most consistent profits and handle the larger order amount while saving distribution costs.

## Data Visualization
Data visualizations are a good way to see the data relative to other data points. In this sample, we are comparing the top 10 states with the most orders. We can see a significant amount of orders in California compared to other places. This allows the business to figure out what is driving such high order count in California and perhaps transfer that model to other places for similar results.  


```python
order_stats_df = order_df['State'].value_counts().head(10)
order_stats_df
order_stats_df.plot(kind='bar', y='Quantity', figsize=(10, 6));
plt.title('Number of Orders by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.show()
```

There is a significant number of orders that is greater in California relative to New York and Texas. Although there are things to consider, such as population size, these states are all a hub for business. There are less sales in the two states outside of California, causing me to believe that this business is based in the west coast and therefore markets to locals as opposed to further out. Perhaps they use word of mouth marketing or something that doesn't travel for far distances.
